### Tidying data
- Importing, understanding, renaming, subsetting, and checking data:

```python
df_vars = ['entity_ID', 'Age','sex','hand oa','oa hip','OAKNEE','cvd','activity','BMI_column','menop']
df_vars_desc = ['ID','Age','Sex','HandOA','HipOA','KneeOA','CVD','Activity','BMI','Menopause']
df_subset = rawdf[df_vars].copy()
df_subset.columns = df_vars_desc
# df_subset.set_index('ID', inplace=True)
df_subset.head()
df_subset.describe()
```

# 
# 

- Reshaping and pivoting tables
    - df.pivot can handle non-numeric data
    - df.melt unpivots from wide to long
    - wide_to_long
    - https://pandas.pydata.org/docs/user_guide/reshaping.html

```python
pivotdf = df_subset.pivot(index='ID',columns='Menopause')
pivotdf.head()
pivotdf.describe()
```

# 
# 

- Combining datsets:
    - pd.concat() can combine multiple frames (with or without keys), will duplicate existing variables
    - join='outer' will take the union of all of the variables
    - join='inner' will join the dfs intersection
    - to append a single row, convert to df and use concat
    - merge() will join them together on key(s)

```python
#Creating new df to merge
new_vars = ['entity_ID','STS']
combiningdf = rawdf[new_vars]
combiningdf = combiningdf.rename(columns={'entity_ID':'ID'})
combiningdf.describe()
```

```python
#concatenating
concatdf = pd.concat([df_subset, combiningdf])
concatdf.head()
concatdf.describe()
```

```python
#merging
mergedf = df_subset.merge(combiningdf, on='ID')
mergedf.head()
mergedf.describe()
```

- Checking for NaNs
    - pd.isna() or.isnull() detects missing values for an object (will return boolean)
    - pd.notna() or .notnull() detects non-missing values
    - Decide what to do with missing values! You can drop arguments or use imputation techniques: https://towardsdatascience.com/imputing-missing-data-with-simple-and-advanced-techniques-f5c7b157fb87

```python
rawdf.isnull().sum() #returns the number of missing values in the dataset
df_subset.isnull().sum()
```
